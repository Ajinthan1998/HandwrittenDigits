```python
import joblib
import cv2
import numpy as np
import time
import pyscreenshot as ImageGrab
from tensorflow.keras.models import load_model

# Load the pre-trained model
model = load_model("C://Users//user//jupyterExamples//Handwritten//bestmodel.h5")
images_folder = "img/"

while True:
    img = ImageGrab.grab(bbox=(60, 170, 400, 500))
    img.save(images_folder + "img.png")
    im = cv2.imread(images_folder + "img.png")
    im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    im_gray = cv2.GaussianBlur(im_gray, (15, 15), 0)

    # Threshold the image
    ret, im_th = cv2.threshold(im_gray, 100, 255, cv2.THRESH_BINARY)
    roi = cv2.resize(im_th, (28, 28), interpolation=cv2.INTER_AREA)

    # Reshape the ROI to match model's input shape
    X = roi.reshape(1, 28, 28, 1)

    # Normalize the input
    X = X.astype('float32') / 255

    # Perform prediction
    predictions = model.predict(X)

    # Get the predicted digit
    predicted_digit = np.argmax(predictions)

    print("Prediction:", predicted_digit)
    cv2.putText(im, "Prediction is: " + str(predicted_digit), (20, 20), 0, 0.8, (0, 255, 0), 2, cv2.LINE_AA)

    cv2.startWindowThread()
    cv2.namedWindow("Result")
    cv2.imshow("Result", im)

    if cv2.waitKey(10) == 13:  # Break the loop when 'Enter' is pressed
        break

cv2.destroyAllWindows()
```
